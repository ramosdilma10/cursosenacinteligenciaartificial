## 8. Criando novas colunas

### Exercício 16
Crie uma coluna chamada `media` com a média entre `nota1` e `nota2`.

### Exercício 17
Crie uma coluna chamada `aprovado`:
- `True` para média maior ou igual a **7**
- `False` caso contrário

### Exercício 18
Crie uma coluna chamada `situacao_faltas`:
- `Alto risco` para mais de 10 faltas
- `Atenção` para 6 a 10 faltas
- `Ok` para até 5 faltas


```python
# Exercício 16 Crie uma coluna chamada media com a média entre nota1 e nota2.
df["media"] = (df["nota1"] + df["nota2"]) / 2
df["media"]

```

```python
# Exercício 17 Crie uma coluna chamada aprovado:
# True para média maior ou igual a 7
# False caso contrário
df["aprovado"] = df["media"] >= 7
df["aprovado"]

```

```python
# Exercício 18 Crie uma coluna chamada situacao_faltas:
# Alto risco para mais de 10 faltas
# Atenção para 6 a 10 faltas
#Ok para até 5 faltas

condicoes = [
    df["faltas"] > 10,
    (df["faltas"] >= 6) & (df["faltas"] <= 10),
    df["faltas"] <= 5
]

escolhas = [
    "Alto risco",
    "Atenção",
    "Ok"
]

df["situacao_faltas"] = np.select(condicoes, escolhas, default=None)
df["situacao_faltas"]
```
